```python
import pandas as pd
base = pd.read_csv('iris.csv')
base.head()
```

```python
X = base.iloc[:, 0:4].values
y = base.iloc[:, 4].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)

from keras.utils import np_utils
y_dummy = np_utils.to_categorical(y)
#setosa      1 0 0
#virginica   0 1 0
#versicolor  0 0 1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_dummy, test_size=0.25, random_state=0)
```

```python
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

clf = Sequential()
clf.add(Dense(units = 4, activation = 'relu', 
                        kernel_initializer = 'random_uniform', input_dim = 4))
clf.add(Dense(units = 4, activation = 'relu', 
                        kernel_initializer = 'random_uniform'))
clf.add(Dense(units = 3, activation = 'softmax'))
clf.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                      metrics = ['categorical_accuracy'])
```

```python
clf.fit(X_train, y_train, batch_size=10, epochs=1000)
```

```python
result = clf.evaluate(X_test, y_test)
```

# Cross Validation

```python
import pandas as pd
base = pd.read_csv('iris.csv')
X = base.iloc[:, 0:4].values
y = base.iloc[:, 4].values

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)

from keras.utils import np_utils
y_dummy = np_utils.to_categorical(y)
#setosa      1 0 0
#virginica   0 1 0
#versicolor  0 0 1
```

```python
def createNetwork():
    import keras
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense

    clf = Sequential()
    clf.add(Dense(units = 4, activation = 'relu', 
                            kernel_initializer = 'random_uniform', input_dim = 4))
    clf.add(Dense(units = 4, activation = 'relu', 
                            kernel_initializer = 'random_uniform'))
    clf.add(Dense(units = 3, activation = 'softmax'))
    clf.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                          metrics = ['categorical_accuracy'])
    return clf
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
clf = KerasClassifier(build_fn=createNetwork, 
                      epochs=1000, 
                      batch_size=10)
results = cross_val_score(estimator=clf, 
                      X=X, y=y, cv=10, 
                      scoring='accuracy')
```

```python
print(results.mean())
print(results.std())
```
