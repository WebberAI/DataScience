```python
from tensorflow.python.framework import ops
ops.reset_default_graph()
```

```python
import pandas as pd
X = pd.read_csv('entradas_breast.csv')
y = pd.read_csv('saidas_breast.csv')
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def buildNetwork():
    import keras
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, Dropout
    clf = Sequential()
    clf.add(Dense(units = 16, 
                  activation = 'relu', 
                  kernel_initializer = 'random_uniform', 
                  input_dim = 30))
    clf.add(Dropout(0.2)) #The Dropout layer randomly sets input units to 0 
    clf.add(Dense(units = 16, 
                  activation = 'relu', 
                  kernel_initializer = 'random_uniform'))
    clf.add(Dropout(0.2))
    clf.add(Dense(units = 1, 
                  activation = 'sigmoid'))
    
    optimizer = keras.optimizers.Adam(lr=0.001, 
                                      decay=0.0001, 
                                      clipvalue=0.5)

    clf.compile(optimizer = optimizer, 
                loss = 'binary_crossentropy',
                metrics = ['binary_accuracy'])
    return clf
```

```python
clf = KerasClassifier(build_fn = buildNetwork, 
                      epochs = 100, 
                      batch_size = 100)

results = cross_val_score(estimator = clf, 
                          X = X, 
                          y = y, 
                          cv = 10, 
                          scoring = 'accuracy')
```

```python
print(results)
print(results.mean())
print(results.std())
```
