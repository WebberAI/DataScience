```python
from tensorflow.python.framework import ops
ops.reset_default_graph()
```

```python
import pandas as pd
X = pd.read_csv('entradas_breast.csv')
y = pd.read_csv('saidas_breast.csv')
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def buildNetwork(optimizer, loos, kernel_initializer, activation, neurons):
    import keras
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, Dropout
    clf = Sequential()
    clf.add(Dense(units = neurons, 
                  activation = activation, 
                  kernel_initializer = kernel_initializer, 
                  input_dim = 30))
    clf.add(Dropout(0.2)) #The Dropout layer randomly sets input units to 0 
    clf.add(Dense(units = neurons, 
                  activation = activation, 
                  kernel_initializer = kernel_initializer))
    clf.add(Dropout(0.2))
    clf.add(Dense(units = 1, 
                  activation = 'sigmoid'))
    
    clf.compile(optimizer = optimizer, 
                loss = loos,
                metrics = ['binary_accuracy'])
    return clf
```

```python
clf = KerasClassifier(build_fn = buildNetwork) # 8hours to run
parameters = {'batch_size': [10, 30], 
              'epochs': [50, 100],
              'optimizer': ['adan', 'sgd'], 
              'loos': ['binary_crossentropy', 'hinge'], 
              'kernel_initializer': ['random_uniform', 'normal'], 
              'activation': ['relu', 'tanh'],
              'neurons': [16, 8]}

from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = clf, 
                           param_grid = parameters, 
                           scoring = 'accuracy', 
                           cv = 5)
grid_search = grid_search.fit(X, y)
```

```python
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
